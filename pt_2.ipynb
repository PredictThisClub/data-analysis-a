```python
from pylab import *
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df = pd.read_csv("abnb.csv").set_index("neighbourhood")
```

```python
x= np.arange(0.0, 2.0, 0.01)
y= sin(2.5*pi*x)
plt.plot([0,0.75,1.5],[-1,1,-1],label= 'first') #line     
plt.plot(x,y,label= 'second')             
plt.title("random")
plt.ylabel("vert")
plt.xlabel("horiz")
plt.legend()
```

```python
x4=df['neighbourhood_group'].unique().tolist()
y41 = df[df['neighbourhood_group']!="Brooklyn"]
y41=y41["price"].mean()
y42 = df[df['neighbourhood_group']!="Manhattan"]
y42=y42["price"].mean()
y43 = df[df['neighbourhood_group']!="Queens"]
y43=y43["price"].mean()
y44 = df[df['neighbourhood_group']!="Staten Island"]
y44=y44["price"].mean()
y45 = df[df['neighbourhood_group']!="Bronx"]
y45=y45["price"].mean()
y4=[y41,y42,y43,y44,y45]
plt.plot(x4,y4)
```

```python
plt.bar(x4,y4,label='bar1',color='b')
```

```python
labels=x4
sizes=y4
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','pink']
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', 
        shadow=True, startangle=140)
plt.axis('equal')
```

```python
labels=x4
sizes=y4
explode = (0.1, 0, 0,0, 0)  # explode 1st slice
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','pink']
plt.pie(sizes,  explode=explode,labels=labels, colors=colors,autopct='%1.1f%%', 
        shadow=True, startangle=180)
plt.axis('equal')
plt.tight_layout()
```

```python
labels=x4
sizes=y4
explode = (0.1, 0, 0,0, 0)  # explode 1st slice
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','pink']
plt.pie(sizes,  explode=explode,labels=labels, colors=colors,autopct='%1.1f%%', 
        shadow=True, startangle=140)
plt.axis('equal')
plt.legend( labels, loc="upper left",shadow=True)
plt.tight_layout()
```

```python
y51 = df[df['neighbourhood_group']!="Brooklyn"]
y51=y51["calculated_host_listings_count"].mean()
y52 = df[df['neighbourhood_group']!="Manhattan"]
y52=y52["calculated_host_listings_count"].mean()
y53 = df[df['neighbourhood_group']!="Queens"]
y53=y53["calculated_host_listings_count"].mean()
y54 = df[df['neighbourhood_group']!="Staten Island"]
y54=y54["calculated_host_listings_count"].mean()
y55 = df[df['neighbourhood_group']!="Bronx"]
y55=y55["calculated_host_listings_count"].mean()
y5=[y51,y52,y53,y54,y55]
```

```python
# Create heatmap
heatmap, xedges, yedges = np.histogram2d(x4,y5, bins=(64,64))
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

# Plot heatmap
plt.clf()
plt.title('heatmap')
plt.ylabel('y')
plt.xlabel('x')
plt.imshow(heatmap, extent=extent)
plt.show()
```

```python
colors = (0,0,0)
area = np.pi*3
plt.scatter(x4, y5, s=area, c=colors, alpha=0.5)
plt.title('housting counts')
plt.xlabel('x')
plt.ylabel('y')
```
